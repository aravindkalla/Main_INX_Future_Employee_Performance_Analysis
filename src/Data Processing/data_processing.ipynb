✅ 2.1 Load and Clean the Data

```python
# data_processing.ipynb

import pandas as pd
import os

# Load the dataset
df = pd.read_excel('../../data/raw/INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xlsx')

# Basic overview
print(df.info())
print(df.describe(include='all'))
display(df.head())

# Check for missing values
print("Missing values per column:\n", df.isnull().sum())

# Fill missing values using forward fill
df = df.ffill()


# Retaining the original EmpDepartment column for visualization or business insights
df_raw = df.copy()  # Keep original for EDA

# Encode categorical variables using one-hot encoding (dropping the first to avoid dummy variable trap)
df_encoded = pd.get_dummies(df, drop_first=True)
#df_encoded = pd.get_dummies(df)

# Create processed data directory if it doesn't exist
os.makedirs('../../data/processed', exist_ok=True)

# Save the final processed DataFrame
df_raw.to_csv('../../data/processed/raw_with_department.csv', index=False)
df_encoded.to_csv('../../data/processed/processed_employee_data.csv', index=False)

print("✅ Processed data saved to '../../data/processed/processed_employee_data.csv'")
```
